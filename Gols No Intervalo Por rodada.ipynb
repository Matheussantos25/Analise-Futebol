```python
import pandas as pd
import numpy as np

# Carregar o arquivo fornecido
file_path_atualizado = 'C:\\Users\\msdof\\OneDrive\\Imagens\\Ligas\\BASE DE DADOS GRANDES\\futpython.xlsx'
dados = pd.read_excel(file_path_atualizado)

# Função para verificar se há 2 gols em um intervalo de 10 minutos e se há mais gols após
def verificar_gols(gols_h, gols_a):
    # Combinar gols de casa e fora em uma lista única
    todos_gols = gols_h + gols_a
    todos_gols = [int(minuto.split('+')[0]) for minuto in todos_gols if minuto != '']
    todos_gols.sort()
    
    # Verificar se há 2 gols em um intervalo de 10 minutos
    for i in range(len(todos_gols) - 1):
        if todos_gols[i + 1] - todos_gols[i] <= 10:
            # Verificar se há mais gols após o segundo gol
            if len(todos_gols) > i + 2:
                return 1
            else:
                return 0
    return 0

# Aplicar a função aos dados
dados['Gols_HT'] = dados.apply(lambda row: verificar_gols(eval(row['Goals_H_Minutes']), eval(row['Goals_A_Minutes'])), axis=1)

# Calcular a probabilidade
total_partidas = len(dados)
partidas_com_gols_adicionais = dados['Gols_HT'].sum()
probabilidade = partidas_com_gols_adicionais / total_partidas

print(f'Probabilidade de ter mais um gol após 2 gols em um intervalo de 10 minutos: {probabilidade:.2%}')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

# Carregar o arquivo fornecido
file_path_atualizado = 'C:\\Users\\msdof\\OneDrive\\Imagens\\Ligas\\BASE DE DADOS GRANDES\\futpython.xlsx'
dados = pd.read_excel(file_path_atualizado)

# Criar a variável alvo
dados['Gol_HT'] = dados['TotalGoals_HT'] > 0

# Verificar as colunas disponíveis
print(dados.columns)

# Unir as colunas HomeTeam e AwayTeam para representar o time
dados['Team'] = dados['Home'] + ' vs ' + dados['Away']

# Verificar a quantidade de rodadas únicas para cada time
rodadas_por_time = dados.groupby('Team')['TotalGoals_HT'].nunique()
print(rodadas_por_time)

# Ajustar a condição para garantir que tenhamos dados suficientes
rodadas_variando = dados.groupby('Team').filter(lambda x: len(x['TotalGoals_HT'].unique()) >= 2)

# Selecionar as features e a variável alvo
X = rodadas_variando[['TotalGoals_HT']]
y = rodadas_variando['Gol_HT']

# Verificar se temos dados suficientes
print(f"Número de amostras: {len(X)}")

# Dividir os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e treinar o modelo
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Fazer previsões
y_pred = model.predict(X_test)
y_pred_prob = model.predict_proba(X_test)[:, 1]

# Avaliar o modelo
print(classification_report(y_test, y_pred))
roc_auc = roc_auc_score(y_test, y_pred_prob)
print(f'AUC: {roc_auc:.2f}')

# Função para simular apostas e calcular lucro/prejuízo
def simular_apostas(y_test, y_pred_prob, threshold=0.5, aposta_inicial=10):
    lucro_prejuizo = 0
    apostas_ganhas = 0
    apostas_perdidas = 0
    
    for i in range(len(y_test)):
        if y_pred_prob[i] > threshold:
            if y_test.iloc[i] == 1:  # aposta ganha
                lucro_prejuizo += aposta_inicial * 1.5  # supondo retorno de 1.5x a aposta
                apostas_ganhas += 1
            else:  # aposta perdida
                lucro_prejuizo -= aposta_inicial
                apostas_perdidas += 1
    
    total_apostas = apostas_ganhas + apostas_perdidas
    win_rate = apostas_ganhas / total_apostas
    roi = (lucro_prejuizo / (total_apostas * aposta_inicial)) * 100
    
    return lucro_prejuizo, win_rate, roi, apostas_perdidas / total_apostas

# Calcular o lucro, win rate e ROI
lucro_prejuizo, win_rate, roi, downgrade = simular_apostas(y_test, y_pred_prob)

print(f'Lucro/Prejuízo: {lucro_prejuizo:.2f}')
print(f'Win Rate: {win_rate:.2f}')
print(f'ROI: {roi:.2f}%')
print(f'Downgrade: {downgrade:.2f}')
```
