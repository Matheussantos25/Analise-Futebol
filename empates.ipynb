```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
dados = pd.read_csv("C:\\Users\\msdof\\OneDrive\\Imagens\\Ligas\\BRA.csv")
dados
```

```python
dados_2023 = dados[dados['Season'] == 2023]
dados_2023
```

```python
dados_2023.loc[:, 'Rodada'] = (dados_2023.index - dados_2023.index[0]) // 10 + 1
dados_2023
```

```python
# Converter a coluna 'HG' para int
dados_2023.loc[:, 'HG'] = dados_2023['HG'].astype(int)

# Converter a coluna 'AG' para int
dados_2023.loc[:, 'AG'] = dados_2023['AG'].astype(int)

# Criar a coluna 'Placar'
dados_2023.loc[:, 'Placar'] = dados_2023['HG'].astype(str) + 'x' + dados_2023['AG'].astype(str)
dados_2023

```

```python
# Criar uma coluna que indica se o jogo foi um empate
dados_2023.loc[:, 'Empate'] = (dados_2023['Res'] == 'D').astype(int)

# Agrupar por 'Rodada' e somar os valores da coluna 'Empate' para obter a quantidade de empates por rodada
empates_por_rodada = dados_2023.groupby('Rodada')['Empate'].sum().reset_index()
empates_por_rodada.columns = ['Rodada', 'Quantidade de Empates']

empates_por_rodada
```

```python
# Exibir as rodadas e a quantidade de empates sem mostrar o índice
print(empates_por_rodada.to_string(index=False))
```

Plotar a frequência de empates por rodada


```python
plt.figure(figsize=(10, 6))
plt.bar(empates_por_rodada['Rodada'], empates_por_rodada['Quantidade de Empates'], color='skyblue')
plt.title('Quantidade de Empates por Rodada - Temporada 2023')
plt.xlabel('Rodada')
plt.ylabel('Quantidade de Empates')
plt.xticks(empates_por_rodada['Rodada'])
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Calcular a quantidade de empates por rodada
dados_2023.loc[:, 'Empate'] = (dados_2023['Res'] == 'D').astype(int)
empates_por_rodada = dados_2023.groupby('Rodada')['Empate'].sum().reset_index()
empates_por_rodada.columns = ['Rodada', 'Quantidade de Empates']

# Encontrar a(s) rodada(s) com o maior número de empates
max_empates = empates_por_rodada['Quantidade de Empates'].max()
rodadas_com_mais_empates = empates_por_rodada[empates_por_rodada['Quantidade de Empates'] == max_empates]

print("Rodada com mais empates:")
print(rodadas_com_mais_empates.to_string(index=False))

# Calcular média, moda e mediana da quantidade de empates por rodada
media = empates_por_rodada['Quantidade de Empates'].mean()
moda = empates_por_rodada['Quantidade de Empates'].mode()[0]  # mode() retorna uma Series, pegamos o primeiro elemento
mediana = empates_por_rodada['Quantidade de Empates'].median()

# Exibir os resultados
print(f"Média de empates por rodada: {media}")
print(f"Moda de empates por rodada: {moda}")
print(f"Mediana de empates por rodada: {mediana}")
```

média, moda, e mediana de empates

```python
# Plotar a quantidade de empates por rodada
plt.figure(figsize=(12, 6))
plt.bar(empates_por_rodada['Rodada'], empates_por_rodada['Quantidade de Empates'], color='skyblue', label='Empates por Rodada')

# Adicionar linhas horizontais para média, moda, e mediana
plt.axhline(media, color='red', linestyle='--', label=f'Média ({media:.2f})')
plt.axhline(moda, color='green', linestyle='-.', label=f'Moda ({moda})')
plt.axhline(mediana, color='purple', linestyle=':', label=f'Mediana ({mediana})')

# Destacar as rodadas com mais empates
for index, row in rodadas_com_mais_empates.iterrows():
    plt.bar(row['Rodada'], row['Quantidade de Empates'], color='gold', label='Rodada com mais empates' if index == rodadas_com_mais_empates.index[0] else "")

plt.title('Análise de Empates por Rodada - Temporada 2023')
plt.xlabel('Rodada')
plt.ylabel('Quantidade de Empates')
plt.xticks(empates_por_rodada['Rodada'])
plt.legend()
plt.grid(axis='y', linestyle='--')

plt.show()
```

Probabilidade de empate geral do cameponato

```python
# Calcular o total de jogos
total_jogos = len(dados_2023)

# Contar quantos jogos terminaram em empate
total_empates = dados_2023[dados_2023['Res'] == 'D'].shape[0]

# Calcular a probabilidade de um jogo terminar em empate
probabilidade_empate = total_empates / total_jogos

# Exibir a probabilidade
print(f"A probabilidade de um jogo terminar em empate no campeonato de 2023 é de {probabilidade_empate:.2f} ou {(probabilidade_empate * 100):.2f}%.")
```

Probabilidade de empate quando ambos times possuem odds >= 1.8

```python
# Filtrar jogos onde as odds da vitória da equipe da casa (PH) E da vitória da equipe visitante (PA) são >= 1.8
jogos_com_odds_altas = dados_2023[(dados_2023['PH'] >= 1.8) & (dados_2023['PA'] >= 1.8)]

# Calcular o total de jogos com odds altas
total_jogos_com_odds_altas = len(jogos_com_odds_altas)

# Contar quantos desses jogos terminaram em empate
total_empates_com_odds_altas = jogos_com_odds_altas[jogos_com_odds_altas['Res'] == 'D'].shape[0]

# Calcular a probabilidade de um jogo terminar em empate sob essas condições
probabilidade_empate_com_odds_altas = total_empates_com_odds_altas / total_jogos_com_odds_altas if total_jogos_com_odds_altas > 0 else 0

# Exibir a probabilidade
print(f"A probabilidade de um jogo terminar em empate, quando ambos os times possuem odds de 1.8 para cima, é de {probabilidade_empate_com_odds_altas:.2f} ou {(probabilidade_empate_com_odds_altas * 100):.2f}%.")
```

Aumentando o filtro, considerando números de empates da rodada

```python
dados_2023.loc[:, 'Empate'] = (dados_2023['Res'] == 'D').astype(int)

# Agrupar por 'Rodada' e somar os valores da coluna 'Empate' para obter a quantidade de empates por rodada
empates_por_rodada = dados_2023.groupby('Rodada')['Empate'].sum().reset_index()

# Filtrar as rodadas com pelo menos 5 empates (considerando que você mencionou 2 mas usou 5 no código)
rodadas_com_pelo_menos_2_empates = empates_por_rodada[empates_por_rodada['Empate'] >= 5]['Rodada']

# Filtrar jogos nessas rodadas e com odds altas para ambas as equipes
jogos_filtrados = dados_2023[(dados_2023['Rodada'].isin(rodadas_com_pelo_menos_2_empates)) & (dados_2023['PH'] >= 1.8) & (dados_2023['PA'] >= 1.8)]

# Calcular a probabilidade de empate nesses jogos filtrados
total_jogos_filtrados = len(jogos_filtrados)
total_empates_filtrados = jogos_filtrados['Empate'].sum()
probabilidade_empate_filtrados = total_empates_filtrados / total_jogos_filtrados if total_jogos_filtrados > 0 else 0

# Exibir a probabilidade
print(f"A probabilidade de empate, após já terem ocorrido 5 empates na rodada e ambos os times possuírem odds de 1.8 para cima, é de {probabilidade_empate_filtrados:.2f} ou {(probabilidade_empate_filtrados * 100):.2f}%.")
```
