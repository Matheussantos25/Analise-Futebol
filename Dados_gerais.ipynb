Dados da Temporada de 2023

```python
dados_2023 = dados[dados['Season'] == 2023]
dados_2023
```

```python
# Copiar o data frame
dados_2023 = dados_2023.copy()
```

Rodadas do campeonato

```python
# Calcular o número total de jogos e número de rodadas
total_jogos = len(dados_2023)
numero_rodadas = total_jogos // 10

# Criar uma coluna para indicar a rodada de cada jogo
# Como Python é baseado em zero, adicionamos 1 para começar a contagem da rodada em 1
dados_2023['Rodada'] = (dados_2023.index - dados_2023.index[0]) // 10 + 1

# Exibir os dados para verificar as rodadas
dados_2023[['Date', 'Home', 'Away', 'Rodada']].head(20)  # Mostrar os 20 primeiros registros para verificar as rodadas
```

```python
#primeiros 20 jogos correspondendo as duas primeiras rodadas
dados_2023[['Date', 'Home', 'Away', 'Rodada']].head(20) 

```

```python
#últimos 20 jogos correspondendo as duas últimas rodadas
dados_2023[['Date', 'Home', 'Away', 'Rodada']].tail(20)

```

Placares do Campeonato Brasileiro em 2023

```python
# Criar uma coluna 'Placar' que combina os gols da casa e os gols do visitante
dados_2023['Placar'] = dados_2023['HG'].astype(str) + '-' + dados_2023['AG'].astype(str)

# Agora, dados_2023 contém uma coluna 'Placar' com os placares das partidas
# Você pode exibir as primeiras linhas para verificar a nova coluna
print(dados_2023[['Date', 'Home', 'Away', 'HG', 'AG', 'Placar']].head())
```

```python
# Garantir que 'HG' e 'AG' são inteiros (para remover qualquer parte decimal)
dados_2023['HG'] = dados_2023['HG'].astype(int)
dados_2023['AG'] = dados_2023['AG'].astype(int)

# Criar uma coluna 'Placar' que combina os gols da casa e os gols do visitante com o formato desejado
dados_2023['Placar'] = dados_2023['HG'].astype(str) + 'x' + dados_2023['AG'].astype(str)

# Agora, dados_2023 contém uma coluna 'Placar' com os placares das partidas no formato desejado
# Você pode exibir as primeiras linhas para verificar a nova coluna
print(dados_2023[['Date', 'Home', 'Away', 'HG', 'AG', 'Placar']].head())
```
